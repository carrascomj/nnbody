```python
import pandas as pd
```

```python
import os
import re

from os.path import pardir, join
```

```python
ROOT = join(pardir)
DATA = join(ROOT, "data", "raw")
PROCESSED = join(ROOT, "data", "processed")
```

```python
seqs_affi = pd.read_excel(join(DATA, "Affimat_AF82422_variant_data_all_AI.xlsx"))
```

```python
seqs_affi.head()
```

```python
pat = re.compile("[L|H]CDR.+ [A-Z]")
```

```python
df = seqs_affi.iloc[:, 14:20]
df.columns = df.columns.str.replace(pat, "")
df["ID"] = seqs_affi["LSP ID"]
df.to_csv(join(PROCESSED, "light_chains.tsv"), sep="\t", index=False)
```

```python
df = seqs_affi.iloc[:, 21:-4]
df.columns = df.columns.str.replace(pat, "")
df["ID"] = seqs_affi["LSP ID"]
df.to_csv(join(PROCESSED, "heavy_chains.tsv"), sep="\t", index=False)
```
